````python
parser = LiteParse(
    ocr_enabled=True,
    tessdata_path="/opt/homebrew/share/tessdata",
    #ocr_language="eng",
)
result = parser.parse("data/attention-is-all-you-need.pdf")
````

````python
from pathlib import Path

pdf_path = "data/attention-is-all-you-need.pdf"

out_dir = Path("output_parsed")
out_dir.mkdir(exist_ok=True)

output_path = out_dir / (Path(pdf_path).stem + ".md")
output_path.write_text(result.text, encoding="utf-8")

print(f"Saved to {output_path}")
````

## Hand written

````python
parser = LiteParse(
    ocr_enabled=True,
    tessdata_path="/opt/homebrew/share/tessdata",
    ocr_language="eng",
)
result = parser.parse("data/handwritten.jpg")
````

````python
parser = LiteParse(
    ocr_enabled=True,
    tessdata_path="/opt/homebrew/share/tessdata",
    ocr_language="eng",
    dpi=300,                    # higher DPI = sharper input for OCR
)
result = parser.parse("data/handwritten.jpg")
print(result.text)
````

## Screenshot

````python
from liteparse import LiteParse
from pathlib import Path
from IPython.display import Image, display  # notebook only

parser = LiteParse(dpi=200)
screenshots = parser.screenshot("data/attention-is-all-you-need.pdf", page_numbers=[1, 2])

out_dir = Path("screenshots")
out_dir.mkdir(exist_ok=True)

for s in screenshots:
    path = out_dir / f"page_{s.page_num:03d}.png"
    path.write_bytes(s.image_bytes)
    display(Image(filename=str(path)))  # remove this line for a .py script
````
